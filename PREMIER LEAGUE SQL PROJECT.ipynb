#         **PREMIER LEAGUE TABLE SQL PROJECT**

```python
import csv, sqlite3
import sqlalchemy
import pandas as pd
import numpy as np
plt.style.use('default')

x = sqlite3.connect('PremierLeague.db')
cursorobj = x.cursor()

%reload_ext sql
%sql sqlite:///PremierLeague.db
    
df = pd.read_csv('allpltables.csv')
df.to_sql('allpltables', x, index=False, if_exists='replace', method='multi')
```

```python
%%sql

SELECT * FROM 'allpltables'
```

# TEAM WITH THE MOST TITLES IN PL HISTORY:

```python
%%sql

SELECT TEAM, count(TEAM) AS 'NUMBER OF TITLES WON'
FROM 'allpltables'
WHERE POSITION = 1
GROUP BY TEAM
ORDER BY count(TEAM) DESC
LIMIT 1;
```

# TEAM WITH THE MOST POINTS IN PL HISTORY:

```python
%%sql

SELECT TEAM, SUM("POINTS") AS 'NUMBER OF POINTS'
FROM 'allpltables'
GROUP BY TEAM
ORDER BY SUM("POINTS") DESC
LIMIT 1;
```

# TEAM WITH THE LEAST AMOUNT OF POINTS IN PL HISTORY:

```python
%%sql

SELECT TEAM, SUM("POINTS") AS 'NUMBER OF POINTS'
FROM 'allpltables'
GROUP BY TEAM
ORDER BY SUM("POINTS")
LIMIT 1;
```

# RECORD FOR MOST POINTS IN  A SINGLE SEASON:

```python
%%sql

SELECT TEAM, MAX(POINTS) AS 'POINTS', SEASON
FROM 'allpltables';
```

# RECORD FOR LEAST POINTS IN  A SINGLE SEASON:

```python
%%sql

SELECT TEAM, MIN(POINTS) AS 'POINTS', SEASON
FROM 'allpltables';
```

# WENT THROUGH AN ENTIRE SEASON UNBEATEN:

# ALL TIME PREMIER LEAGUE TABLE:

```python
%%sql

SELECT
TEAM, 
SUM(MATCHES) AS MATCHES,
SUM(WINS) AS WINS, 
SUM(DRAWS)AS DRAWS, 
SUM(LOSSES) AS LOSSES, 
SUM("GOALS FOR") AS "GOALS FOR", 
SUM("GOALS AGAINST") AS "GOALS AGAINST", 
SUM("GOAL DIFFERENCE") AS "GOAL DIFFERENCE", 
SUM(POINTS) AS POINTS

FROM 'allpltables'
GROUP BY TEAM
ORDER BY SUM(POINTS) DESC;
```
